```python
with open("sample_data/clean_json.json", "r") as read_file:
    data = json.load(read_file)

print(data)
```

```python
def update(a, b):
    a.update(b)
    return a

df = pd.read_json("sample_data/clean_json.json", lines=True).to_dict('r')
pd.json_normalize([update(y, x) for x in df for y in x.pop('track')['features']])
```

```python
#normalize step 1: get it all

df= pd.json_normalize(data, record_path=['resources'], meta=['contentVersion', 'parameters', 'variables', 'outputs'])
print(df)
```

```python
df.info()
```

```python
parameter_df = df["parameters"]
print(parameter_df)
```

```python
parameter_df.to_json(orient='columns')
```

```python
parameter_df= pd.json_normalize(parameter_df)
print(parameter_df)
```

```python
variables_df = df["variables"]
variables_df.to_json(orient='columns')
variables_df = pd.json_normalize(variables_df)
print(variables_df)

outputs_df = df["outputs"]
outputs_df.to_json(orient='columns')
outputs_df = pd.json_normalize(outputs_df)
print(outputs_df)
```

```python
df_merged = pd.concat([df, parameter_df, variables_df, outputs_df], axis=1)



print(df_merged)
```
